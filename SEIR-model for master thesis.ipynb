```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import sympy as sym
from sympy import *
from scipy import integrate
from scipy.integrate import simps
init_printing()
```

USA

```python
def SEIR(X,t, mu, beta, x, e, sigma, gamma):    
    S = X[0]
    E = X[1]  
    I = X[2]
    R = X[3]
    
    dSdt = mu - mu*S - beta*S*I*(1-e*x) - x*e*S
    dEdt = beta*S*I*(1-e*x) - (sigma+mu)*E
    dIdt = sigma*E - (gamma + mu)*I
    dRdt = gamma*I - mu*R + x*e*S
    return [dSdt, dEdt, dIdt, dRdt]

t = np.linspace(0,200,200)
ini = [0.97286, 0.004525, 0.004525, 0.01809] 

def inter(mu, beta, x, e, sigma, gamma):
    X = odeint(SEIR, ini, t, args=(mu, beta, x, e, sigma, gamma))
    S = X[:,0]
    E = X[:,1]
    I = X[:,2]
    R = X[:,3]
    plt.clf()
    plt.plot(t,S, label = "Suceptible", color="dimgray")
    plt.plot(t,E, label = "Exposed", color="darksalmon")
    plt.plot(t,I, label = "Infected", color="rosybrown")
    plt.plot(t,R, label = "Recovered", color="burlywood")    
    plt.legend()
    plt.show()
    
    print('Max number of infected: ',np.max(I))
    ind = np.where(I == np.max(I))
    print('On day number',ind[0][0])
        
    
mu, beta, x, e, sigma, gamma = 0.00003178, 0.462, 0.002, 0.52, (1/11.5), 0.0686
inter(mu, beta, x, e, sigma, gamma)
```
